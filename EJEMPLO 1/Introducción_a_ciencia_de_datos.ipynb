```python
# Libreria seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Figura 1

sns.scatterplot(data=df, x="2G", y="4G",hue='Empresa')

plt.show()
```

```python
#Figura 2

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

# Agrupar por Región y sumar la cantidad de antenas 3G

region_3g = df.groupby("Region")["3G"].sum().reset_index()

# Crear el gráfico con Seaborn

plt.figure(figsize=(10, 6))

sns.barplot(y="Region", x="3G", data=region_3g, palette="viridis")

plt.title("Cantidad de Antenas 3G por Región")

plt.xticks(rotation=45)

plt.show()
```
