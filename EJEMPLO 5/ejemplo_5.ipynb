```python
import pandas as pd
df = pd.read_excel("Ejemplo5.xlsx")
```

```python
df.info()
```

```python
df["MONTO"] = df ["CANTIDAD"]*df["PRECIO"]
df.info()
```

```python
df
```

```python
#CONVERTIR A DATATIME

df['FECHA'] = pd.to_datetime(df['FECHA'])

df.info()
```

```python
ventas_mensuales = (

  df

  .groupby(df['FECHA'].dt.to_period('M'))

  .agg({'MONTO': 'sum'})

  .reset_index()

  .rename(columns={'FECHA': 'PERIODO', 'MONTO': 'VENTAS'})

)

ventas_mensuales.set_index('PERIODO',inplace=True)
```

```python
import matplotlib.pyplot as plt

ventas_mensuales.plot()

plt.show()
```
