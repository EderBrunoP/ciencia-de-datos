```python
import plotly.express as px

# Obtener los conteos y renombrar columnas
df_altocol = df['Altocolesterol'].value_counts().reset_index()
df_altocol.columns = ['Altocolesterol', 'Cantidad']

# Crear gráfico
fig = px.bar(df_altocol,
             x='Altocolesterol', y='Cantidad',
             labels={'Altocolesterol': '¿Colesterol alto?', 'Cantidad': 'Cantidad'},
             title="Colesterol alto",
             color='Altocolesterol')

# Cambiar etiquetas del eje x
fig.update_xaxes(tickvals=[0, 1], ticktext=["No", "Sí"])
fig.show()

```

```python
import plotly.express as px

df_actividad = df['Actividadfísica'].value_counts().reset_index()
df_actividad.columns = ['Actividadfísica', 'Cantidad']

# Crear gráfico de barras
fig = px.bar(df_actividad,
             x='Actividadfísica', y='Cantidad',
             labels={'Actividadfísica': '¿Hace actividad física?', 'Cantidad': 'Cantidad'},
             title="Actividad física",
             color='Actividadfísica')

# Etiquetas personalizadas en el eje X
fig.update_xaxes(tickvals=[0, 1], ticktext=["No", "Sí"])
fig.show()

```

```python
import plotly.express as px

# Contar y renombrar columnas
df_fumadoras = df['Fumador'].value_counts().reset_index()
df_fumadoras.columns = ['Fumador', 'Cantidad']

# Crear gráfico de barras
fig = px.bar(df_fumadoras,
             x='Fumador', y='Cantidad',
             labels={'Fumador': '¿Fumadora?', 'Cantidad': 'Cantidad'},
             title="Fumadoras",
             color='Fumador')

# Etiquetas del eje X (asumiendo 0 = No, 1 = Sí)
fig.update_xaxes(tickvals=[0, 1], ticktext=["No", "Sí"])
fig.show()
```
