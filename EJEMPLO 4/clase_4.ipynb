```python
#Librería de vectores

import numpy as np

# Es la librería que selecciona los datos para entrenamiento y prueba

from sklearn.model_selection import train_test_split

# Es la librería del modelo clasificación k vecinos

from sklearn.neighbors import KNeighborsClassifier

# Es la librería de métricas

from sklearn.metrics import classification_report, confusion_matrix
```

```python
#Plantilla

df = pd.read_excel("Ejemplo-kvecinos.xlsx")

variablesX = ['Edad','Ingresos','Estudios']

etiquetaY = 'Etiqueta'

test_porcentaje = 0.10

k_vecinos = 8
```

```python
#1 Datos

X = df[variablesX]

y = df[etiquetaY]

#2 Entrenamiento

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_porcentaje, random_state=42)

modelo = KNeighborsClassifier(n_neighbors=k_vecinos)

modelo.fit(X_train, y_train)
```

```python
X_test
```

```python
#3 Evaluación

y_pred = modelo.predict(X_test)

print("Reporte de classification:")

print(classification_report(y_test, y_pred))

plt.figure(figsize=(6,6))

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d")

plt.xlabel('Predicciones')

plt.ylabel('Real')

plt.show()
```

```python
#Producción (cuidado con la etiqueta 1==soltero,0==casado)

#varibales para nuevos datos

nuevos_datos = pd.DataFrame({'Edad': [1],'Ingresos': [1], 'Estudios': [4]})

prediccion = modelo.predict(nuevos_datos)

print("Predicción:", prediccion)

if prediccion == 1:

  print("Soltero")

else:

  print("Casado")
```
