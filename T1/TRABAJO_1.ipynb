```python
# prompt: media, mediana y moda, con
# interpretaciones detalladas en español los comenrarios

import pandas as pd

# importando libreria panda
#leer el archivo
df=pd.read_excel('Trabajo1.xlsx')

# información dataframe
df.info()
df.describe()

# Calcular la media, mediana y moda para cada columna numérica
for column in df.select_dtypes(include=['number']):
  media = df[column].mean()
  mediana = df[column].median()
  moda = df[column].mode()

  print(f"\nAnálisis descriptivo para la columna: {column}")
  print(f"Media: {media}")
  print(f"Mediana: {mediana}")
  print(f"Moda: {moda.iloc[0]}") #Accede al primer elemento de la Serie moda

  # Interpretación detallada
  print("\nInterpretación:")
  print(f"La media de {column} es {media}, representando el promedio de los valores.")
  print(f"La mediana de {column} es {mediana}, indicando el valor central de los datos ordenados.")
  print(f"La moda de {column} es {moda.iloc[0]}, mostrando el valor que más se repite en la columna.")

  #Comparacion entre media y mediana para identificar posibles asimetrías
  if media > mediana:
    print("La media es mayor que la mediana, lo que sugiere una posible asimetría positiva (cola derecha) en la distribución de los datos.")
  elif media < mediana:
    print("La media es menor que la mediana, lo que sugiere una posible asimetría negativa (cola izquierda) en la distribución de los datos.")
  else:
    print("La media es igual a la mediana, lo que sugiere una distribución simétrica de los datos.")


```

```python
# prompt:  rango, desviación estándar y cuartiles.

import pandas as pd

# importando libreria panda
#leer el archivo
df=pd.read_excel('Trabajo1.xlsx')

# información dataframe
df.info()
df.describe()

# Calcular la media, mediana, moda, rango, desviación estándar y cuartiles para cada columna numérica
for column in df.select_dtypes(include=['number']):
  media = df[column].mean()
  mediana = df[column].median()
  moda = df[column].mode()
  rango = df[column].max() - df[column].min()
  desviacion_estandar = df[column].std()
  cuartiles = df[column].quantile([0.25, 0.5, 0.75])

  print(f"\nAnálisis descriptivo para la columna: {column}")
  print(f"Media: {media}")
  print(f"Mediana: {mediana}")
  print(f"Moda: {moda.iloc[0] if not moda.empty else 'No hay moda'}") # Maneja el caso sin moda
  print(f"Rango: {rango}")
  print(f"Desviación estándar: {desviacion_estandar}")
  print(f"Cuartiles:\n{cuartiles}")

  # Interpretación detallada (puedes agregar más interpretaciones según tus necesidades)
  print("\nInterpretación:")
  print(f"La media de {column} es {media}, representando el promedio de los valores.")
  print(f"La mediana de {column} es {mediana}, indicando el valor central de los datos ordenados.")
  print(f"La moda de {column} es {moda.iloc[0] if not moda.empty else 'No hay moda'}, mostrando el valor que más se repite.")
  print(f"El rango de {column} es {rango}, mostrando la dispersión de los datos.")
  print(f"La desviación estándar de {column} es {desviacion_estandar}, indicando la dispersión típica de los datos alrededor de la media.")
  print(f"Los cuartiles de {column} son:\nPrimer cuartil (Q1): {cuartiles[0.25]}\nSegundo cuartil (Q2 o mediana): {cuartiles[0.5]}\nTercer cuartil (Q3): {cuartiles[0.75]}")

  #Comparacion entre media y mediana para identificar posibles asimetrías
  if media > mediana:
    print("La media es mayor que la mediana, lo que sugiere una posible asimetría positiva (cola derecha).")
  elif media < mediana:
    print("La media es menor que la mediana, lo que sugiere una posible asimetría negativa (cola izquierda).")
  else:
    print("La media es igual a la mediana, lo que sugiere una distribución simétrica.")
```

```python
# prompt: Interpretan y contextualizan los resultados estadísticos claramente y con relevancia.

import pandas as pd

# Lee el archivo Excel (asegúrate de que 'Trabajo1.xlsx' esté en el mismo directorio o proporciona la ruta completa)
try:
    df = pd.read_excel('Trabajo1.xlsx')
except FileNotFoundError:
    print("Error: El archivo 'Trabajo1.xlsx' no se encontró. Por favor, verifica la ruta del archivo.")
    exit() # Termina la ejecución si el archivo no se encuentra

# Información general del DataFrame
print(df.info())
print(df.describe())


# Análisis descriptivo para cada columna numérica
for column in df.select_dtypes(include=['number']):
    media = df[column].mean()
    mediana = df[column].median()
    moda = df[column].mode()
    rango = df[column].max() - df[column].min()
    desviacion_estandar = df[column].std()
    cuartiles = df[column].quantile([0.25, 0.5, 0.75])

    print(f"\nAnálisis descriptivo para la columna: {column}")
    print(f"Media: {media:.2f}")  # Formato a dos decimales
    print(f"Mediana: {mediana:.2f}")
    print(f"Moda: {moda.iloc[0] if not moda.empty else 'No hay moda'}")
    print(f"Rango: {rango:.2f}")
    print(f"Desviación estándar: {desviacion_estandar:.2f}")
    print(f"Cuartiles:\nQ1: {cuartiles[0.25]:.2f}\nQ2 (Mediana): {cuartiles[0.5]:.2f}\nQ3: {cuartiles[0.75]:.2f}")

    # Interpretación contextualizada
    print("\nInterpretación:")
    print(f"La media de {column} es {media:.2f}, indicando el valor promedio de la variable.  ")
    print(f"La mediana de {column} es {mediana:.2f}, representando el valor central de los datos. ")

    if not moda.empty:
      print(f"La moda es {moda.iloc[0]},  el valor que más se repite en el conjunto de datos. Esto puede indicar una tendencia o concentración de los datos alrededor de ese valor.")
    else:
      print("No hay moda, lo que significa que no hay valores repetidos con mayor frecuencia en la columna.")


    print(f"El rango de {column} es {rango:.2f}, indicando la diferencia entre el valor máximo y el mínimo, mostrando la dispersión de los datos. ")
    print(f"Una desviación estándar de {desviacion_estandar:.2f} indica que los valores se desvían, en promedio, {desviacion_estandar:.2f} unidades de la media.")
    print("Los cuartiles proporcionan información sobre la distribución de los datos y la presencia de valores atípicos.  ")

    # Comparación media-mediana para asimetría
    if abs(media - mediana) / desviacion_estandar > 0.2:  #Criterio basado en la desviación estándar
        if media > mediana:
            print("La media es significativamente mayor que la mediana, lo que sugiere una posible asimetría positiva (cola derecha).")
        else:
            print("La media es significativamente menor que la mediana, lo que sugiere una posible asimetría negativa (cola izquierda).")
    else:
        print("La media y la mediana son similares, lo que sugiere una distribución aproximadamente simétrica.")
```

```python
# prompt: Escogen gráficos adecuados (barras, dispersión, boxplot, etc.) para el tipo de datos y análisis.

import matplotlib.pyplot as plt
import seaborn as sns

# ... (Your existing code) ...


# Gráficos para cada columna numérica
for column in df.select_dtypes(include=['number']):
    plt.figure(figsize=(8, 6))  # Ajusta el tamaño de la figura si es necesario

    # Histograma para visualizar la distribución de los datos
    plt.subplot(2, 2, 1)
    sns.histplot(df[column], kde=True)  # kde=True agrega una curva de densidad
    plt.title(f'Histograma de {column}')

    # Boxplot para identificar valores atípicos y la distribución de los cuartiles
    plt.subplot(2, 2, 2)
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot de {column}')

    # Diagrama de dispersión (si tienes otra columna numérica para comparar)
    # Reemplaza 'otra_columna' con el nombre de una columna numérica
    if len(df.select_dtypes(include=['number']).columns) > 1:
        otra_columna = df.select_dtypes(include=['number']).columns[1] # Selecciona la segunda columna numérica como ejemplo
        if column != otra_columna : # Evita graficar contra sí misma
          plt.subplot(2, 2, 3)
          sns.scatterplot(x=df[column], y=df[otra_columna])
          plt.title(f'Diagrama de dispersión de {column} vs {otra_columna}')

    # Gráfico de violín para una representación alternativa de la distribución
    plt.subplot(2, 2, 4) #Posicion del gráfico en el subplot
    sns.violinplot(y=df[column])
    plt.title(f'Gráfico de violín de {column}')


    plt.tight_layout()  # Ajusta automáticamente los subplots para evitar superposiciones
    plt.show()
```

```python
# prompt: Gráficos claros, con ejes bien etiquetados, títulos, leyendas y unidades.

import matplotlib.pyplot as plt
# ... (Your existing code) ...

# Gráficos para cada columna numérica
for column in df.select_dtypes(include=['number']):
    plt.figure(figsize=(10, 8))  # Ajusta el tamaño de la figura

    # Histograma
    plt.subplot(2, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribución de {column}')
    plt.xlabel(column)  # Etiqueta del eje x
    plt.ylabel('Frecuencia')  # Etiqueta del eje y

    # Boxplot
    plt.subplot(2, 2, 2)
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot de {column}')
    plt.ylabel(column)  # Etiqueta del eje y


    # Diagrama de dispersión (si hay otra columna numérica)
    if len(df.select_dtypes(include=['number']).columns) > 1:
        other_column = df.select_dtypes(include=['number']).columns[
            (df.select_dtypes(include=['number']).columns.get_loc(column) + 1) % len(df.select_dtypes(include=['number']).columns) ] # Selecciona la siguiente columna numérica
        if column != other_column:
          plt.subplot(2, 2, 3)
          sns.scatterplot(x=df[column], y=df[other_column])
          plt.title(f'{column} vs {other_column}')
          plt.xlabel(column)  # Etiqueta del eje x
          plt.ylabel(other_column)  # Etiqueta del eje y

    # Gráfico de violín
    plt.subplot(2, 2, 4)
    sns.violinplot(y=df[column])
    plt.title(f'Gráfico de violín de {column}')
    plt.ylabel(column)  # Etiqueta del eje y

    plt.tight_layout()
    plt.show()
```

```python
# prompt: grafico de cantidad total de alquileres de "Distrito" por "Año"

import matplotlib.pyplot as plt
# ... (Your existing code) ...

# Agrupar por 'Año' y 'Distrito' y contar los alquileres
alquileres_por_distrito_año = df.groupby(['Año', 'Distrito'])['Distrito'].count().reset_index(name='Cantidad de Alquileres')

# Crear el gráfico
plt.figure(figsize=(12, 6))
sns.barplot(x='Año', y='Cantidad de Alquileres', hue='Distrito', data=alquileres_por_distrito_año)
plt.title('Cantidad total de alquileres por Distrito y Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Alquileres')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor legibilidad
plt.tight_layout()
plt.show()
```

```python
# prompt: grafico de cantidad total para indicar el top 10 de "Distrito" por "Año"

import matplotlib.pyplot as plt
# ... (Your existing code) ...

# Agrupar por 'Año' y 'Distrito' y contar los alquileres
alquileres_por_distrito_año = df.groupby(['Año', 'Distrito'])['Distrito'].count().reset_index(name='Cantidad de Alquileres')

# Obtener el top 10 de distritos por año
top_10_distritos = alquileres_por_distrito_año.groupby('Año').apply(lambda x: x.nlargest(10, 'Cantidad de Alquileres')).reset_index(drop=True)


# Crear el gráfico para el top 10
plt.figure(figsize=(16, 8))
sns.barplot(x='Año', y='Cantidad de Alquileres', hue='Distrito', data=top_10_distritos)
plt.title('Top 10 de distritos por cantidad de alquileres para cada año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Alquileres')
plt.xticks(rotation=45)
plt.legend(title='Distrito', bbox_to_anchor=(1.05, 1), loc='upper left') # Ajusta la leyenda
plt.tight_layout()
plt.show()
```
