```python
# Final
# Quiero estudiar sexo,edad,altura y peso + id_persona
df = pd.read_excel("Ejemplo2.xlsx")
df = df[["id_persona",'sexo','edad','Altura','Peso']]
df = df.dropna()
df = df.drop_duplicates(subset=['id_persona'])
df.info()
```

```python
# estad√≠sticas en general
df[["Altura","edad","Peso"]].describe()
```

```python
# Comando mean
df[["Altura","edad","Peso"]].mean()
```

```python
# Comando loc, aplica un filtro
# Promedio por mujeres
df_mujeres = df.loc[df["sexo"]=="F"]
df_mujeres.describe()
```

```python
# Comando loc, aplica un filtro
# Promedio por hombres
df_hombres = df.loc[df["sexo"]=="M"]
df_hombres.describe()
```

```python
# Comando groupby
# los promedios de las variables edad, peso y altura por sexo
df[["edad","Peso","Altura","sexo"]].groupby(["sexo"]).mean()
```

```python
# Comando groupby
# los medianas de las variables edad, peso y altura por sexo
df[["edad","Peso","Altura","sexo"]].groupby(["sexo"]).median()
```

```python
# Comando groupby
# las sumas de las variables edad, peso y altura por sexo
df[["edad","Peso","Altura","sexo"]].groupby(["sexo"]).sum()
```
