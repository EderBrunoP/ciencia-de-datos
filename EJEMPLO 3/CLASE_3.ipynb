# ***Repaso***

```python
import pandas as pd
df = pd.read_excel('Ejemplo3.xlsx')
df.info()
```

```python
# Eliminar duplicados
df = df.drop_duplicates(subset=	['Codigo'])
df.info()
```

```python
# Eliminar vacios
df = df.dropna()
df.info()
```

```python
# estadisticas basicas
df.describe()
```

```python
# El precio promedio es de 50.48
# El precio mínimo es de 2.40
# El precio máximo es de 99.11
# El cuartil 1 del precio es 22.797500 (el 25% de los precios son menores a 27.797500)
```

```python
# Herramienta groupby
df[['Costos_de_fabricacion','Nombre_del_proveedor']].groupby('Nombre_del_proveedor').mean()
# el costo promedio de fabricación del proveedor 1 es de 21.04
```

```python
#Herramienta groupby

df[['Transportistas_de_envio','Tiempos_de_envio']].groupby('Transportistas_de_envio').mean()

#El tiempo promedio de envio del proveedor DHL es 4.5135 dias
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.boxplot(data=df, x='Nombre_del_proveedor', y='Plazo_de_entrega')

plt.show()

# x:string

# y:numerica
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.histplot(data=df, x='Precio')

# x:numerica

plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='Costos_de_fabricacion',y='Precio')

plt.show()

# x:numerica

# y:numerica
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.pairplot(data=df[['Precio','Costos_de_fabricacion','Tiempos_de_envio','Transportistas_de_envio']], hue='Transportistas_de_envio' )

plt.show()

# varias variables : numericas

# una variable string para HUE
```

```python
#graficos mas estilizados

import plotly.express as px

fig = px.scatter(df,

         x='Costos_de_fabricacion',

         y='Precio',

         color='Nombre_del_proveedor',

         size='Tiempos_de_envio',

         facet_col='Transportistas_de_envio')

fig.show()
```

```python
fig = px.histogram(

    df,

    x="Precio",

    color="Tipo_de_producto",  # Diferenciar por tipo de producto

    nbins=30,

    title="Distribución de Precios por Tipo de Producto",

    hover_data=["Codigo", "Numero_de_productos_vendidos", "Nombre_del_proveedor"],  # Info extra al pasar el mouse

    opacity=0.75

)



fig.update_layout(

    template="plotly_white",  # Tema visual

    xaxis_title="Precio del Producto",

    yaxis_title="Cantidad de Productos",

    legend_title="Tipo de Producto",

    bargap=0.05,

    font=dict(family="Arial", size=14),

    title_font=dict(size=20),

    hoverlabel=dict(bgcolor="white", font_size=13, font_family="Arial"),

    margin=dict(t=60, b=50, l=50, r=50)

)

fig.update_traces(marker_line_width=1, marker_line_color="black")  # Bordes negros a las barras

fig.write_html("histograma_precios.html")
```
